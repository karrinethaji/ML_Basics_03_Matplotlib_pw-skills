## Matplotlib Library:

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

**Matplotlib is the visualization library. we have a variety of visualization libraries available in present market like seaborn, plotly, Bokeh etc.,**

**In matplotlib we plot the graphs using x and y-axis** like a stock market graphs ( on x-axis (time) and y-axis (price of share))

```python
# import matplotlib library

import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# create the two varaiables x and y

x=np.random.rand(50)
y=np.random.rand(50)
```

```python
print(x)
```

```python
print(y)
```

```python
# Lets plot x and y using matplotlib

plt.plot(x,y)
```

```python
# plotting scatter plot

plt.scatter(x,y)
```

```python
# we can provide the label names for x and y-axis and title

plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is the graphg betweeen x and y")
```

```python
# we can change the colour of graphs

plt.scatter(x,y,c="green")# c='g' also works also you can provide the html color codes in the field c
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is the graphg betweeen x and y")
```

# HTML Colour codes

#website: https://html-color.codes/



```python
# we can change the colour of graphs using html color code

plt.scatter(x,y,c="#6383be")# you can provide the html color codes in the field c
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is the graphg betweeen x and y")
```

```python
# we can also change the internsity of colour using the variable alpha (less alpha value then color intensity is also less)

print(plt.scatter(x,y,c="r",alpha=0.5))
```

```python
print(plt.scatter(x,y,c='r',alpha=1))
```

```python
# If you want to represent the plotting data with "x" instaed of circle then we need to use "marker"

print(plt.scatter(x,y,c='r',alpha=1,marker="x"))
```

```python
print(plt.scatter(x,y,c='r',alpha=1,marker="*"))
```

```python
# If you want to print the graph with grid 

plt.scatter(x,y,c="#6383be")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is the graphg betweeen x and y")
plt.grid()
```

```python
# we can change the figure size
plt.figure(figsize=(10,5))
plt.scatter(x,y,c="#6383be")# you can provide the html color codes in the field c
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is the graphg betweeen x and y")
```

```python
# Bar plot

a= list("abcde") # ['a', 'b', 'c', 'd', 'e']

b= np.random.rand(5)

print(b)
```

```python
plt.bar(a,b)
plt.show()
```

```python
plt.bar(a,b,color="#00ff00")
plt.title("Representing bar plot with categorical data on x-axis and numerical data on y-axis")
plt.xlabel("Representing the categorical data")
plt.ylabel("Representing the numerical data")
```

```python
plt.figure(figsize=(8,5))
plt.bar(a,b,color="g")
plt.title("Representing bar plot with categorical data on x-axis and numerical data on y-axis")
plt.xlabel("Representing the categorical data")
plt.ylabel("Representing the numerical data")
```

```python
# Horizontal bar graph

plt.figure(figsize=(8,5))
plt.barh(a,b,color="g")
plt.title("Representing bar plot with categorical data on x-axis and numerical data on y-axis")
plt.xlabel("Representing the categorical data")
plt.ylabel("Representing the numerical data")
```

```python
# Histogram graph ( Histogram will plot the count of occurences of the data)

x=[11,11,11,33,33,22,33,22,33,33,33,44,55,55,55,66,66,100,33,22,22]
```

```python
plt.figure(figsize=(8,5))
plt.hist(x)# learn about the bins
plt.title("Representing hist plot with count of occurences of data")
plt.xlabel("Representing the x-axis data")
plt.ylabel("Representing the count of data")
plt.grid()
```

```python
# Saving the figure

plt.figure(figsize=(8,5))
plt.hist(x)# learn about the bins
plt.title("Representing hist plot with count of occurences of data")
plt.xlabel("Representing the x-axis data")
plt.ylabel("Representing the count of data")
plt.grid()
plt.savefig("histplot.png")
```

```python
# Plotting 3D graph

x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
```

```python
# 3D Plot

x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig= plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z)
```

```python
# 3D Plot

x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig= plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z,color="r",marker="x")
plt.xlabel("this is x-axis")
plt.ylabel("this is y-axis")
plt.show()
```
