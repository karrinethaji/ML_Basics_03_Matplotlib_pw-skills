## Matplotlib:

**Matplotlib** is a plotting library for the python programming language and its numerical mathmatics extension numpy. It provide an object oriented API for embedding plots into applications using general-purpose GUI toolkits using Tkinter,Wxpython etc.,

**Advantages of using matplotlib:**

1. Generally easy to get started 
2. Easy to plot the simple plots
3. Supports for the custom labels and texts
4. Great control of every element in a figure
5. High quality of output in many formats

```python
# import the matplotlib package

from matplotlib import pyplot as plt

# import matplotlib.pyplot as plt

%matplotlib inline

# use of %matplotlib inline: In general in matplotlib we use plt.show() to get the output, but if you declare the %matplotlib inline
# then you don't need to specify the plt.show() for every figure to get the output

# import numpy package as well

import numpy as np
```

```python
# Let's plot some basic plots

x=np.arange(0,10)
y=np.arange(11,21)

plt.scatter(x,y)
plt.xlabel("this is x-axis")
plt.ylabel("this is y-axis")
plt.title("This is x vs y graph")
plt.savefig("nethajitest.png")# saving the created graph with name "nethajitest.png"
```

```python
x=np.arange(0,10)
y=np.arange(11,21)

# we can change the colour of the graphs
plt.scatter(x,y,color="green",marker="*") # In the color field we can also use the HTML color codes 
plt.xlabel("this is x-axis")
plt.ylabel("this is y-axis")
plt.title("This is x vs y graph")
plt.savefig("nethajitest.png")# saving the created graph with name "nethajitest.png"
```

```python
# In matplotlib the lineplot can be plotted using the function plot()
x=np.arange(0,10)
y=np.arange(11,21)
plt.plot(x,y,color="red",marker="*")
plt.show() # comment this line of code and observe what happens
```

```python
x=np.arange(0,10)
y=x*x

plt.plot(x,y,"--r",marker="*")
```

```python
x=np.arange(0,10)
y=x*x

plt.plot(x,y,"--r",marker="*",linewidth=3,markersize=10)
plt.show()
```

```python
# creating the subplots

fig,ax=plt.subplots(2,2) # 2 rows and 2 columns == 4 graphs

x=np.linspace(-1,1,101)
y1=x**2
y2=x**3
y3=x**4
y4=x**5


ax[0,0].plot(x,y1) # grpah of 0th index row 0th index column
ax[0,1].plot(x,y2) # grpah of 0th index row 1st index column
ax[1,0].plot(x,y3,color="green") # grpah of 1st index row 0th index column
ax[1,1].plot(x,y4,"r--") # grpah of 1st index row 1st index column
plt.show()
```
